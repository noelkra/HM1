## Rechnerarithmetik und Zahlensysteme

### Zahlensysteme und Konvertierungen
Konvertieren von Dezimalzahlen in andere Zahlensysteme und umgekehrt.

```python
def convert_to_base(n, base):
    """Convert a decimal number n to a given base and return it as a string."""
    if n == 0:
        return "0"
    digits = []
    while n:
        digits.append(int(n % base))
        n //= base
    return ''.join(str(x) for x in digits[::-1])
```

```python
def convert_from_base(s, base):
    """Convert a number in string format from a given base to decimal."""
    return sum(int(digit) * (base ** idx) for idx, digit in enumerate(reversed(s)))
```

### Maschinengenauigkeit
Bestimmung der Maschinengenauigkeit (Epsilon) für Gleitkommazahlen sowie berechnung von Rundungsfehlern.

```python
def absoluterFehler(wahr, gemessen):
    """Calculate the absolute error between the true value and the measured value."""
    return abs(wahr - gemessen)

def relativerFehler(wahr, gemessen):
    """Calculate the relative error between the true value and the measured value."""
    return abs((wahr - gemessen) / wahr) if wahr != 0 else float('inf')
```

```python
def maschinengenauigkeit():
    """Calculate the machine epsilon for floating point numbers."""
    epsilon = 1.0
    while (1.0 + epsilon) != 1.0:
        epsilon /= 2.0
    return epsilon * 2.0
```

### Konditionszahl und Konditionierung
Berechnung der Konditionszahl einer Funktion und Analyse der Konditionierung eines Problems.

```python
def konditionszahl(f, df, x):
    """Calculate the condition number of a function f at point x."""
    f_x = f(x)
    df_x = df(x)
    return abs(x * df_x / f_x) if f_x != 0 else float('inf')
```
